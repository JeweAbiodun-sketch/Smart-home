```python

"""
HomeGuard Security System Simulator
Author: [Olawuyi Adetunji]
Description: A smart home monitoring system that processes sensor readings
             and triggers alerts for security, safety, and comfort issues.
"""
 
import random
import datetime
 
# System configuration
HOME_MODES = ["HOME", "AWAY", "SLEEP"]
ALERT_SEVERITIES = ["LOW", "MEDIUM", "HIGH", "CRITICAL"]
 
# Current system state
current_mode = "AWAY"



# ----------------------------
# SENSOR CLASS
# ----------------------------
class Sensor:
    def __init__(self, name, sensor_type):
        self.name = name
        self.type = sensor_type
        self.value = None

    # Generate fake sensor readings
    def generate_reading(self):
        if self.type == "motion":
            self.value = random.choice([True, False])

        elif self.type == "door":
            self.value = random.choice([True, False])

        elif self.type == "temperature":
            self.value = random.randint(20, 110)

        elif self.type == "smoke":
            self.value = random.choice([True, False])

        return self.value


# ----------------------------
# HOME SECURITY SYSTEM
# ----------------------------
class HomeGuardSystem:

    def __init__(self):
        # Dictionary for system state
        self.system_state = {
            "mode": "home",  # home or away
            "alerts": []
        }

        # Dictionary for sensors
        self.sensors = {
            "motion": Sensor("Living Room Motion", "motion"),
            "door": Sensor("Front Door", "door"),
            "temperature": Sensor("Thermostat", "temperature"),
            "smoke": Sensor("Smoke Detector", "smoke")
        }

    # Change system mode
    def set_mode(self, mode):
        self.system_state["mode"] = mode
        print(f"\nSystem mode set to: {mode.upper()}")

    # Check all sensors
    def process_sensors(self):
        self.system_state["alerts"] = []  # reset alerts

        motion = self.sensors["motion"].generate_reading()
        door = self.sensors["door"].generate_reading()
        temperature = self.sensors["temperature"].generate_reading()
        smoke = self.sensors["smoke"].generate_reading()

        print("\n--- Sensor Readings ---")
        print("Motion:", motion)
        print("Door Open:", door)
        print("Temperature:", temperature)
        print("Smoke:", smoke)

    
        # CONDITIONAL LOGIC (if/else)
        # SECURITY

        if self.system_state["mode"] == "away":
            if motion:
                self.system_state["alerts"].append("Motion detected while away!")
            if door:
                self.system_state["alerts"].append("Door opened while away!")

        if motion and door:
            self.system_state["alerts"].append("Multiple sensors triggered!")

        # SAFETY
        if temperature < 35:
            self.system_state["alerts"].append("Temperature below 35°F!")

        elif temperature > 95:
            self.system_state["alerts"].append("Temperature above 95°F!")

        if smoke:
            self.system_state["alerts"].append("Smoke detected!")

        # COMFORT
        if self.system_state["mode"] == "home":
            if temperature < 65 or temperature > 75:
                self.system_state["alerts"].append("🌡️ Temperature outside comfort range!")

        # Show alerts
        self.display_alerts()

    # Display alerts
    def display_alerts(self):
        print("\n--- Alerts ---")

        if not self.system_state["alerts"]:
            print("No alerts. System normal.")
        else:
            for alert in self.system_state["alerts"]:
                print(alert)


# ----------------------------
# MAIN PROGRAM
# ----------------------------
if __name__ == "__main__":

    system = HomeGuardSystem()

    # Change to away mode to test security
    system.set_mode("away")

    # Run simulator 3 times
    for time in range(3):
        system.process_sensors()
        time = datetime.datetime.now()
print(time)
```

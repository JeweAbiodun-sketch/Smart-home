```python
# This class is the blueprint for our smart home system
class HomeGuard:

    # This runs automatically when we create the system
    def __init__(self):
        # The house starts in "home" mode by default
        self.mode = "home"

        # Comfortable temperature range (min, max)
        self.comfort_range = (65, 75)

    # This function changes the house mode (home or away)
    def set_mode(self, mode):
        self.mode = mode
        print(f"\nSystem mode set to: {self.mode.upper()}")

    # This function checks all sensors and decides if alerts are needed
    def process_sensors(self, motion=False, door=False, temperature=70, smoke=False):

        # Create an empty list to store warning messages
        alerts = []

        # -------------------------
        # SECURITY CHECKS
        # -------------------------

        # If house is in away mode, any motion or door activity is suspicious
        if self.mode == "away":
            if motion:
                alerts.append("🚨 Motion detected while away!")
            if door:
                alerts.append("🚨 Door opened while away!")

        # If both motion and door happen together, it could be a break-in
        if motion and door:
            alerts.append("🚨 Multiple sensors triggered! Possible break-in!")

        # -------------------------
        # SAFETY CHECKS
        # -------------------------

        # Temperature too low (risk of frozen pipes)
        if temperature < 35:
            alerts.append("❄️ Temperature below 35°F (Frozen pipe risk)")

        # Temperature too high (equipment could fail)
        elif temperature > 95:
            alerts.append("🔥 Temperature above 95°F (Equipment failure risk)")

        # Smoke detection means possible fire
        if smoke:
            alerts.append("🔥 Smoke detected! Possible fire!")

        # -------------------------
        # COMFORT CHECKS
        # -------------------------

        # Only care about comfort if people are home
        if self.mode == "home":
            if temperature < self.comfort_range[0] or temperature > self.comfort_range[1]:
                alerts.append("🌡️ Temperature outside comfort range (65-75°F)")

        # Return all alerts found
        return alerts


# MAIN PROGRAM (Simulation)

# Create the smart home system
system = HomeGuard()

# Set house to away mode
system.set_mode("away")

# Simulate sensor readings
alerts = system.process_sensors(
    motion=True,
    door=True,
    temperature=30,
    smoke=False
)

# Print all alerts
for alert in alerts:
    print(alert)
```
